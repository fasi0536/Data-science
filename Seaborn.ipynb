```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: student marks
marks = [35, 45, 50, 50, 55, 60, 60, 60, 65, 70, 75, 80]

# Create a distribution plot
sns.distplot(marks,bins=5)

# Add labels and title
plt.title("Distribution of Student Marks")
plt.xlabel("Marks")
#plt.ylabel("Frequency")

# Show the plot
plt.show()
```

```python

marks = [35, 40, 40, 45, 50, 50, 50, 55, 60, 65, 70, 70]

sns.histplot(marks, bins=5, kde=False)
plt.title("Student Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")  # Y-axis
plt.show()
```

### LM Plot (Linear model plot)

```python
import pandas as pd
```

```python
# Sample Data
data = pd.DataFrame({
    'Hours': [1, 2, 3, 4, 5, 6],
    'Marks': [40, 50, 55, 65, 70, 80]
})

# lmplot = scatter plot + best-fit line
sns.lmplot(x='Hours', y='Marks', data=data)
plt.title("Study Hours vs Marks")
plt.grid()
plt.show()
```
